```python
import pandas as pd
pd.set_option('display.max_rows', 500)
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_notebook
```

```python
with open("NetflixViewingHistory_krika.csv") as file:
    df = pd.read_csv(file)
    
def label_type(row):
    if ":" in row['Title']:
        return "Series"
    else:
        return "Movie"

df['Type'] = df.apply(lambda row: label_type(row), axis=1)
df['Title'] = df['Title'].str.split(":").str[0]
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Type'] == "Series"]
#dates = pd.date_range(start=df['Date'].min(), end= df['Date'].max())

with open('titles-times.csv') as file:
    df3 = pd.read_csv(file, header=None)
df3 = df3.drop(df3.columns[0], axis=1)
df3.columns = ['Title','Length']
join = pd.merge(left=df, right=df3)
join['Date'] = pd.to_datetime(join['Date'])
join.set_index('Date', inplace=True)

df4 = join.groupby(join.index)['Length'].sum()
df4 = df4/60
df4.sort_index(inplace=True)
```

```python
def yearly_plot(year=float):
    df_yearly = df4.where(df4.index.year==year)

    plot_yearly = figure(title= f"My Netflix Analysis {year}", x_axis_label="day", y_axis_label="hours watched", x_axis_type= "datetime", plot_width=1100)
    plot_yearly.vbar(width=1, x=df_yearly.index, top=df_yearly)
    show(plot_yearly)

    dfweek = df_yearly.groupby(df_yearly.index.isocalendar().week).mean()
    plot_week = figure(title= f"Week avrg {year}", x_axis_label="week", y_axis_label="avg hours watched per day", plot_width=1100)
    plot_week.vbar(width=.5, x=dfweek.index, top=dfweek)
    show(plot_week)

    sum = df_yearly.sum()/24
    avg = df_yearly.mean()
    print(f"sum {year}: ",  sum, f"average {year}: ", avg)

```

```python
yearly_plot(2020)
```

```python
yearly_plot(2018)
```

```python
yearly_plot(2017)
```
